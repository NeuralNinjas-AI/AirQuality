## News

```python
df = pd.read_csv('datasets/news/merged_data.csv')
```

```python
df
```

```python
stop_words = set(stopwords.words("english"))
def preprocess_text(text):
    text = text.lower()
    text = ''.join([char for char in text if char not in string.punctuation])
    text = ' '.join([word for word in text.split() if word not in stop_words])
    return text
```

```python
df['Snippet'] = df['Snippet'].apply(preprocess_text)
```

```python
NEWS = " ".join(snippet for snippet in df['Snippet'])
```

```python
df['Snippet']
```

```python
wordcloudNews = WordCloud(width=1000, height=800, background_color="white", colormap='viridis').generate(NEWS)
```

```python
plt.figure(figsize=(20, 8))
plt.imshow(wordcloudNews, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
tokens = word_tokenize(NEWS)

# Filter out stop words
filtered_tokens = [word for word in tokens if word.lower() not in stop_words]

# Calculate word frequencies for the filtered tokens
freq_dist = FreqDist(filtered_tokens)

# Select the most common words (e.g., top 20)
top_wordsNEWS = freq_dist.most_common(20)
```

```python
plt.figure(figsize=(12, 6))
plt.bar([word[0] for word in top_wordsNEWS], [word[1] for word in top_wordsNEWS])
plt.xlabel("Words")
plt.ylabel("Frequency")
plt.title("Top 20 Word Frequencies 2019 NEWS")
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.show()
```

```python
import nltk
import pandas as pd
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Download the vader_lexicon, which contains the rules for the Vader SentimentAnalyzer. VADER is a lexicon and rule-based sentiment analysis tool that is specifically attuned to sentiments expressed in social media.
nltk.download('vader_lexicon')

# Initialize the VADER sentiment intensity analyzer
sid = SentimentIntensityAnalyzer()

# Assume df is your DataFrame and 'text_column' is the column with the text
df['sentiment'] = df['Snippet'].apply(lambda text: sid.polarity_scores(text))

# Convert the sentiment scores dictionary into columns
df = pd.concat([df.drop(['sentiment'], axis=1), df['sentiment'].apply(pd.Series)], axis=1)
```

```python
df
```

```python
# Create a histogram of the compound sentiment scores
plt.hist(df['compound'], bins=20, color='skyblue')
plt.title('Sentiment Analysis')
plt.xlabel('Compound Score')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
from wordcloud import WordCloud

# Pie chart of sentiment categories
df['sentiment_category'] = pd.cut(df['compound'], bins=3, labels=['negative', 'neutral', 'positive'])
sentiment_counts = df['sentiment_category'].value_counts()
plt.pie(sentiment_counts, labels=sentiment_counts.index, autopct='%1.1f%%')
plt.title('Sentiment Categories')
plt.show()
```

```python
# Word clouds of positive and negative words
positive_words = ' '.join(df[df['compound'] > 0.3]['Snippet'])
neutral_words = ' '.join(df[(df['compound'] >= -0.3) & (df['compound'] <= 0.3)]['Snippet'])
negative_words = ' '.join(df[df['compound'] < -0.3]['Snippet'])

wordcloud_positive = WordCloud().generate(positive_words)
wordcloud_neutral = WordCloud().generate(neutral_words)
wordcloud_negative = WordCloud().generate(negative_words)

plt.imshow(wordcloud_positive, interpolation='bilinear') 
plt.title('Positive Words')
plt.axis("off")
plt.show()

plt.imshow(wordcloud_neutral, interpolation='bilinear')
plt.title('Neutral Words')
plt.axis("off")
plt.show()

plt.imshow(wordcloud_negative, interpolation='bilinear')
plt.title('Negative Words')
plt.axis("off")
plt.show()
```

```python
# Time series plot of sentiment
df['date'] = pd.to_datetime(df['MatchDateTime'])
df.set_index('date', inplace=True)
df['compound'].resample('1W').mean().plot() # take the mean for each week
plt.title('Sentiment Over Time')
plt.xlabel('Date')
plt.ylabel('Compound Score')
plt.show()
```
