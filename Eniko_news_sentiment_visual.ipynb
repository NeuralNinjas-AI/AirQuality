```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('./sentiment.csv')
```

```python
sentiment_counts = df['Sentiment'].value_counts()
```

```python
sentiment_counts.plot(kind='bar')
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()
```

```python
sentiment_percentages = (sentiment_counts / sentiment_counts.sum()) * 100

# Create a bar plot to visualize the sentiment distribution
sentiment_percentages.plot(kind='bar')
plt.title('Sentiment Distribution (Percentage)')
plt.xlabel('Sentiment')
plt.ylabel('Percentage')
plt.show()
```

```python
plt.pie(sentiment_percentages, labels=sentiment_percentages.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that the pie chart is circular.
plt.title('Sentiment Distribution (Percentage)')
plt.show()
```

```python
import plotly.express as px
```

```python
sentiment_labels = {
    0: 'admiration',
    1: 'amusement',
    2: 'anger',
    3: 'annoyance',
    4: 'approval',
    5: 'caring',
    6: 'confusion',
    7: 'curiosity',
    8: 'desire',
    9: 'disappointment',
    10: 'disapproval',
    11: 'disgust',
    12: 'embarrassment',
    13: 'excitement',
    14: 'fear',
    15: 'gratitude',
    16: 'grief',
    17: 'joy',
    18: 'love',
    19: 'nervousness',
    20: 'optimism',
    21: 'pride',
    22: 'realization',
    23: 'relief',
    24: 'remorse',
    25: 'sadness',
    26: 'surprise',
    27: 'neutral',
}
```

```python
sentiment_percentages.index = sentiment_percentages.index.map(sentiment_labels)
```

```python
fig = px.treemap(
    names=sentiment_percentages.index,
    parents=[''] * len(sentiment_percentages),  # All at the root level
    values=sentiment_percentages.values,
    title='Sentiment Distribution (Percentage)'
)

fig.show()
```
